# Documenting Linear Algebra

Typesetting cheat sheet for quantum relevant formulas.

### Qubits

ket zero: $\left | 0 \right> = \begin{bmatrix}
1\\
0
\end{bmatrix}$
ket one: $\left | 1 \right> = \begin{bmatrix}
0\\
1
\end{bmatrix}$
ket psi: $\left | \psi \right> = \alpha \left | 0 \right> + \beta \left | 1 \right> = \begin{bmatrix}
\alpha\\
\beta
\end{bmatrix}$, for $\alpha, \beta \in \mathbb{C}$ and $\vert\alpha\vert^2 + \vert\beta\vert^2 = 1$

### Complex Conjugate

$\alpha = a + ib \rightarrow \alpha^* = a - ib$

### Conjugate Transpose

$\bra{\psi} = \ket{\psi}^\dagger = \begin{bmatrix}
\alpha_1^* & \alpha_2^* & \dots & \alpha_d^*\\  
\end{bmatrix}$, with $\psi = \begin{bmatrix}
\alpha_1\\
\alpha_2\\
\vdots\\
\alpha_d\\
\end{bmatrix}$

$\ket{\psi} = \bra{\psi}^\dagger$

### Matrices 

$H = \frac{1}{\sqrt2}\begin{bmatrix}
1 & 1\\
1 & -1
\end{bmatrix}$

### Tensor Product

tensor product of quantum states: $\left | 0 \right> \otimes \left | 1 \right> = \left | 01 \right>$

scaler can be moved around: given kets $\left | \phi \right>, \left | \psi \right>, \left | \rho \right>$, and $\left | \sigma \right>$, and $r \in \mathbb{R}$, then $r(\left | \phi \right> \otimes \left | \psi \right>) = (r\left | \phi \right>) \otimes \left | \psi \right> = \left | \phi \right> \otimes (r\left | \psi \right>)$

### Inner Product 

$\braket{\psi}{\phi} = \sum_{i=1}^{d} \alpha_i^* \beta_i$, with $\phi = \begin{bmatrix}
\beta_1\\
\beta_2\\
\vdots\\
\beta_d\\
\end{bmatrix}$

### Outer Product

$\ket{\psi} \bra{\phi} = \begin{bmatrix}
\alpha_1\\
\alpha_2\\
\vdots\\
\alpha_d\\
\end{bmatrix} \begin{bmatrix}
\beta_1^* & \beta_2^* & \dots & \beta_d^*\\  
\end{bmatrix} = 
\begin{bmatrix}
\alpha_1 \beta_1^* & \dots & \alpha_1 \beta_d^* \\
\alpha_2 \beta_1^* & \dots & \alpha_2 \beta_d^* \\
\vdots & \ddots & \vdots \\
\alpha_d \beta_1^* & \dots & \alpha_d \beta_d^* \\
\end{bmatrix}$

### Entanglement

two entangled states using H and C_NOT (one of the Bell states): $\Phi^+ = \frac{1}{\sqrt2} (\left | 00 \right> + \left | 11 \right>)$

### Universal Boolean Oracle

$U_f \left | x_1x_2...x_ny \right> = \left | x_1x_2...x_n f(x_1x_2...x_n)\oplus y\right>$

### Phase Oracle $O$

$O_f\ket{x} = \begin{cases}\ket{x} \text{ if} f(x) = 0 \\ - \ket{x} \text{ if} f(x) = 1 \end{cases}$

## Pennylane Basics

*source: csc_486_586_lecture_material(10)*

```python
import math

import pennylane as qml
```

A qml.device is needed to create a quantum circuit. A QNode represents a quantum node, but has to perform a measurement and return the result.

```python
dev = qml.device('default.qubit', wires=1, shots=1024)


@qml.qnode(dev)  # quantum circuit
def circuit():
    qml.Hadamard(wires=0)  # index of qubit
    return qml.probs(wires=0)  # probabilities for number of shots
```

Return the probablities after measurement with 1024 shots

```python
print(f"Probabilities: {circuit()}")
```

Return statevector instead of probabilities

```python
dev = qml.device('default.qubit', wires=2)  # remove shots parameter


@qml.qnode(dev)  # quantum circuit
def circuit():
    qml.Hadamard(wires=0)  # index of qubit
    qml.CNOT(wires=[0, 1])
    return qml.state()  # return state instead of prob
```

```python
print(f"State: {circuit()}")
```

```python
c = 0
d = 1
```

Initialize the quantum circuit with two qubits.

```python
alice = QuantumRegister(1, "Alice")
bob = QuantumRegister(1, "Bob")
protocol = QuantumCircuit(alice, bob)

display(protocol.draw(output="mpl", style="iqp"))
```

Create Bell state $\left |\Phi^+\right>$

```python
# Prepare ebit used for superdense coding
protocol.h(alice)
protocol.cx(alice, bob)
protocol.barrier()

display(protocol.draw(output="mpl", style="iqp"))
```

Map Bell state basis back to computational basis to be able to measure d and c

```python
# Bob's actions
protocol.cx(alice, bob)
protocol.h(alice)
protocol.measure_all()  # only for testing purposes

display(protocol.draw(output="mpl", style="iqp"))
```

```python
result = Sampler().run(protocol).result()
statistics = result.quasi_dists[0].binary_probabilities()

for outcome, frequency in statistics.items():
    print(f"Measured {outcome} with frequency {frequency}")

display(plot_histogram(statistics))
```

# Grover's Algorithm

Grover's Algorithm finds a target string $x_0$ of length $n$ in a search space $S = \{0, 1, ..., N-1\}$ and oracle function $f:\{0,1, ..., N-1\} \rightarrow \{0,1\}$ where $f(x) = 1$ for exactly one element, the target string $x_0 \in S$, and $f(x) = 0$ otherwise.

Classically, we would have to evaluate $f(x)$ for every element $x$ in the search space $S$, leaving us with a runtime of $O(|S|)$.

Thanks to superposition, only $n = log_2N$ qubits are needed to encode the entire search space $S$, resulting in a quantum search time complexity of $O(\sqrt{N})$. 

![image](imgs/grovers_algorithm.png)
*source: Slides Lecture 7 Grover*

The algorithm starts with equal superposition, in order to model all $2^n$ possible inputs. The phase oracle $O_f$ tags the amplitude of the target element $x_0$ with a negative phase. Then, the diffusion operator $D$ increases its amplitude by inversion over the mean. Together, applying $O_f$ and $D$ form one iteration.

My original misconception was that the algorithm finds one qubit $q_x$ that resembles the target value of either 0 or 1. Actually, it maximizes the probability of receiving the target vector $x_0$ when measuring $n$ qubits. It was important to realize that the number of qubits $n$, is not equal to the number of amplitudes $N$, and that each amplitude describes the probability of one of the elements from $S$.

The following is an implementation, in order to visualize the update rule of Grover's algorithm. Implementing the update rule gave me more intuition on the meaning of amplitudes, as well as the relation to their probabilities, as well as on the intuition behind superposition. 

Update rule that Diffusion Operator $D$ implements: $\alpha'_x = 2a - \alpha_x$ with $a = \frac{1}{N} \sum_{x \in S} \alpha_x$, the mean of the amplitudes, and $N=2^n$, the number of elements in $S$, the input set, and $n$ the number of qubits.

```python
from matplotlib import animation
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import seaborn as sns
import random
import math
import numpy as np
import pandas as pd
```

```python
# rule used inverse over mean of amplitudes
def update_rule(alpha, a):
    return 2 * a - alpha
```

Initialize the amplitudes. For simplicity, define the number of qubits $n$, rather than the size of the search space $N$. Change $n$ to visualize Grover's algorithm for a different number of qubits.

```python
n = 5  # number of qubits 
N = 2 ** n  # number of elements in S

iterations = list()  # store results after applying diffusion operator

# initialize the amplitudes with superposition state
amplitudes = [1 / math.sqrt(N)] * N
iterations.append(amplitudes.copy())  # add initial amplitudes
```

Now the iterations begin. The phase oracle $O_f$ is simulated by adding a negative phase to an amplitude randomly. The ideal amount of iterations that maximize the probability of measuring the target string are $k = \frac{\pi}{4}\sqrt{N}$

```python
k = (math.pi / 4) * math.sqrt(N)
target_idx = random.randrange(N)

# start the iterations
for i in range(round(k)):  # round k to nearest integer
    # simulation of a phase oracle that tags one element of the amplitudes with a negative phase
    amplitudes[target_idx] = - amplitudes[target_idx]

    # apply update rule
    a = np.mean(amplitudes)  # compute mean of amplitudes
    amplitudes = [update_rule(alpha, a) for alpha in amplitudes]

    iterations.append(amplitudes.copy())  # add after one iteration
```

Visualize the amplitudes at each step of the algorithm

```python
pd.DataFrame(iterations).transpose().plot.bar()
```

For $n = 5$, the algorithm runs for 5 iterations. Each iteration, the amplitudes of the target string $x_0$ increases, while all other amplitudes slowly decrease. 

The following code visualizes the same process with an animation

```python
%matplotlib notebook

fig = plt.figure(figsize=(8, 6))
axes = fig.add_subplot(1, 1, 1)
axes.set_ylim(-1, 1)

subtitles = [f"x_{i}" for i in range(N)]  # create list of titles  


def animate(i):
    plt.bar(subtitles, iterations[i])


plt.title("Grover's Algorithm", color="blue")
ani = FuncAnimation(fig, animate)
```

# Quantum Fourier Transform (QFT)

The QFT is the quantum counterpart of the Discrete Fourier Transform (DFT) algorithm. The DFT transforms a signal into the frequency domain, whereas the QFT transforms the basis from computational into Fourier basis. (More about [Changing Basis](#Changing-Basis-in-Linear-Algebra). The Hadamard gate is the equivalent to the QFT for the 1-qubit case:

\begin{equation}
H(\alpha \ket{0} + \beta \ket{1}) =  \alpha \ket{+} + \beta \ket{-}
\end{equation}

The Hadamard gate can also be seen as adding a phase of $e^{\frac{2\pi i\psi}{2}}$. This makes sense when rewriting the equation of applying the Hadamard gate:

\begin{equation}
H\ket{\psi} =  \frac{1}{\sqrt{2}} (\ket{0} + e^{\frac{2\pi i\psi}{2}} \ket{1})
\end{equation}.

The added phase results to $1$ if $\psi = 0$, which creates the $\ket{+}$ state. The phase results in $-1$, if $\psi = 1$, creating the $\ket{-}$ state.

For the $n$-qubit case, the formula therefore is the following:

\begin{equation}
QFT\ket{\psi} = \frac{1}{\sqrt{N}}(\ket{0}+e^{\frac{2\pi i\psi}{2^1}}\ket{1}) \otimes (\ket{0}+e^{\frac{2\pi i\psi}{2^2}}\ket{1}) \otimes \dots \otimes (\ket{0}+e^{\frac{2\pi i\psi}{2^n}}\ket{1})
\end{equation}

, what essentially means that we are adding the phase $e^{\frac{2\pi i\psi}{2^k}}$ to every $\ket{\psi_k}$.

## Gate for QFT

For realizing a quantum circuit for the $n$-qubit QFT case, we need the $U_{ROT_k} = \begin{bmatrix} 1 & 0 \\ 0 & e^{\frac{2\pi i\psi}{2^k}}\end{bmatrix}$ gate, which rotates the qubit around the z-axis by the factor $\frac{2\pi}{2^k}$.

## Applications

In quantum computing, QFT has multiple applications. It is used for determining a phase of an eigenvalue in the Quantum Phase Estimation (QPE) algorithm. More precisely, the inverse of the QFT, $QFT^\dagger$, is one of the building blocks of QPE. It is used to convert a state in Fourier basis back to the computational basis, to be able to make a measurement. 

Furthermore, QPE is used in Shor's algorithm, which makes QFT also necessary for it. It is also used for computing discrete logarithms via a variant of Shor's algorithm.

Just as DFT is used for signal processing, QFT can be used for tasks like signal filtering, spectral analysis, and pattern recognition.

Another application area of QFT is Quantum Simulation, which implies simulating the dynamics of quantum systems, such as molecules, materials, and complex physical systems.

Last but not least, QFT can be used for Quantum Machine Learning (QML) as well. For example, it is used in quantum principal component analysis (PCA) and quantum clustering. Especially PCA is used to reduce the input dimension of training data, which improves the efficiency of certain machine learning tasks.

# Quantum Phase Estimation (QPE)

In your own words, describe the Quantum Phase Estimation (QPE) algorithm. How is QPE used in quantum computing?

QPE is an algorithm used to estimate the global phase of a quantum state. This is neccessary, because the global phase does not show up when measuring a quantum state, only the relative phase might (more on [Relative and Global Phases](#Relative-and-Global-Phases)). Figuring out the global phase is equivalent to finding the eigenvalue $e^{2\pi i\theta}$ of a quantum gate $U$ corresponding to a given eigenvector $\ket{\psi}$. When applying $U$ on one of its eigenvectors $\ket{\psi}$, it adds the phase $e^{2\pi i\theta}$, which resembles the eigenvalue:

\begin{equation}
U\ket{\psi} = e^{2\pi i\theta}\ket{\psi}
\end{equation}

(more on [eigenvalues](#Eigenvalues-and-Eigenstates-in-Quantum-Computing))

> *Reminder*: a matrix $U$ is unitary, if its inverse equals its conjugate transpose: $U^\dagger U = I$, and all quantum gates are unitary matrices

To be able to estimate the global phase of other states than eigenvectors of the unitary $U$, we utilize *phase kickback*. We prepare an auxiliary qubit in the eigenstate of the unitary operator, while we prepare the control wires in equal superposition. Then, when we apply the controlled version of our unitary, which kicks back the eigenvalue corresponding to the eigenstate prepared in the target wire to the control wires.

1) Initial state for 1 control and 1 target wire:
\begin{equation}
    \ket{0} \otimes \ket{\psi}
\end{equation}
2) Transform control wires into equal superposition:
\begin{equation}
    H(\ket{0}) \otimes \ket{\psi} = \frac{1}{\sqrt{2}}(\ket{0} + \ket{1}) \otimes \ket{\psi} = \frac{1}{\sqrt{2}}(\ket{0\psi} + \ket{1\psi})
\end{equation}
3) Apply controlled Unitary U using $U\ket{\psi} = e^{2\pi i\theta}\ket{\psi}$
\begin{equation}
    \frac{1}{\sqrt{2}}(\ket{0\psi} + \ket{1}U\ket{\psi}) = \\ \frac{1}{\sqrt{2}}(\ket{0\psi} + \ket{1}e^{2\pi i\theta}\ket{\psi}) = \\
    \frac{1}{\sqrt{2}}(\ket{0} + e^{2\pi i\theta}\ket{1}) \otimes \ket{\psi}
\end{equation}

To be able to measure the phase, we transform our state in Fourier basis back to the computational basis by applying the inverse of the QFT, $QFT^\dagger$. For the 1-qubit case, this is the Hadamard gate:

4) Transform back to computational basis:
\begin{equation}
    H(\frac{1}{\sqrt{2}}(\ket{0} + e^{2\pi i\theta}\ket{1})) \otimes \ket{\psi} = \\
    \frac{1}{\sqrt{2}}(\frac{1}{\sqrt{2}}(\ket{0} + \ket{1}) + e^{2\pi i\theta} \frac{1}{\sqrt{2}}(\ket{0} - \ket{1})) \otimes \ket{\psi} = \\
    \frac{1}{2}((1 + e^{2\pi i\theta})\ket{0} + (1 - e^{2\pi i\theta})\ket{1}) \otimes \ket{\psi}
\end{equation}

using definition of Hadamard:
\begin{equation}
    H(\alpha\ket{0} + \beta\ket{1}) = \frac{1}{\sqrt{2}}((\alpha + \beta) \ket{0} + (\alpha - \beta) \ket{1})
\end{equation}


#### Simple QPE Circuit for Unitary Gate Z

$Z = \begin{bmatrix}1 & 0 \\ 0 & -1\end{bmatrix}$

Eigenstates of $Z$: $\ket{1}, \ket{0} = \ket{\psi}$

```python
control_wire = QuantumRegister(1, "control")  # this will be initialized in equal superposition
target_wire = QuantumRegister(1, "target")  # this will be the eigenstate
c_bit = ClassicalRegister(1)
qpe = QuantumCircuit(control_wire, target_wire, c_bit)

qpe.h(control_wire)  # prepare in equal superposition
qpe.x(target_wire)  # prepare in eigenstate ket 1
qpe.cz(control_wire, target_wire)  # apply controlled unitary (z)
qpe.h(control_wire)  # inverse qft for one qubit
qpe.measure(control_wire, 0)

display(qpe.draw())
```

```python
sampler = Sampler()
results = sampler.run(qpe).result()

results.quasi_dists
```

The measurement of the control wire results in 1. The result of the circuit measurement is encoding the phase angle $\theta$ using fractional binary:

\begin{equation}
\theta = 0.\theta_1 \theta_2 \dots \theta_n = \frac{\theta_1}{2} + \frac{\theta_2}{2^2} + \dots + \frac{\theta_n}{2^n} = \sum^n_{k=1} \frac{\theta_k}{2^k}
\end{equation}

In our example, $\theta_1 = 1$, therefore $\theta = 0.\theta_1 = 0.1 = \frac{\theta_1}{2} = \frac{1}{2}$, with the corresponding phase $e^{2\pi i\theta} = e^{2\pi i \frac{1}{2}} = 1$. This resembles our estimated eigenvalue, corresponding to the eigenstate $\ket{1}$. 


#### t-bit precision

This example works well for the $Z$-gate, since its eigenvalues $1$ and $-1$ can be represented with the phase angles $1$ and $0$ respectively, which are the two possible measurement outcomes for our 1-qubit state. Looking at another example unitary, $S = \begin{bmatrix}1&0\\0&e^{2 \pi i \frac{1}{4}}\end{bmatrix}$, the eigenvalue corresponding to eigenstate $\ket{1}$ is $\frac{1}{4}$. This value can not be represented as a fractional binary of only one bit, but needs two: $\theta = 0.\theta_1\theta_2 = 0.01 = \frac{1}{4}$. Therefore, increasing the number of qubits also increases the number of phase angles we can precisely represent. This is called *t-bit precision*, where $t$ resembles the number of estimation wires. 

#### Apply U $2^k$ times

If we add more estimation wires, we also need to kick back the estimated phase onto each of them separately. We also need to make sure that in the end, all estimation qubit measurements form the fractional binary representation of the phase angle. Therefore, we apply the controlled unitary number of estimation wires $t$ times, with each of the estimation wires being the control wire once.  More precisely, we need to apply $U$ more than one time on each estimation wire. We apply it $2^k$ times in order to move the decimal in our fractional binary $k$ to the right. We start with $k = t-1$ for the first qubit, to add a phase angle of $0.\theta_t$, continue with $k=t-2$, to add the phase angle $0.\theta_{t-1}\theta_t$, and so on, until $k=0$ for the last qubit $t$. Basically, we are reverse engineering this state in Fourier basis, to be able to extract the desired state  $\ket{\theta_1\theta_2\dots\theta_t}$ in the next step.

#### Apply $QFT^\dagger$

The state that we created in the previous step is represented in the Fourier basis. In order to measure our state in the computational basis, we apply the inverse of the $QFT$ to transform the state back to the computational basis.

Finally, we can measure our estimation wires, where each wire $t$ represents $\theta_t$.

#### QPE Circuit with 2-bit precision

*source: qpe-fillable.ipynb assignment 3*

In this example we estimate the eigenvalue for the given gate $S = \begin{bmatrix}1&0\\0&e^{2 \pi i \frac{1}{4}}\end{bmatrix}$, where the eigenvalue corresponding to eigenstate $\ket{1}$ is $\frac{1}{4}$.

```python
dev = qml.device("default.qubit", wires=3)
estimation_wires = [0, 1]
target_wires = [2]

def prepare_eigenvector():
    qml.PauliX(wires=target_wires)
    

def estimate_phase(probs, n_estimation_wires):
    # estimate the phase angle by transforming index corresponding to highest probability from 
    # fractional binary to decimal
    return np.argmax(probs) / 2 ** n_estimation_wires

@qml.qnode(dev)
def qpe(unitary):
    """ Estimate the phase for a given unitary.
    Args:
        unitary (array[complex]): A unitary matrix.
    Returns:
        probs (array[float]): Probabilities on the estimation wires.
    """
    prepare_eigenvector() # prepare eigenvector ket 1
    qml.QuantumPhaseEstimation(unitary, target_wires, estimation_wires)  # Qiskit QPE template
    
    return qml.probs(estimation_wires)


U = qml.S.compute_matrix()
probs = qpe(U)
print("Estimated phase angle theta =", estimate_phase(probs, len(estimation_wires)))
```

# Decomposition Strategies for Hybrid Quantum-Classical Systems

The currently availabe quantum computers, so-called noisy-intermediate-scale-quantum (NISQ) devices, are limited in their abilities. This will change in the future, as a lot of research is conducted on improving the current state of the art. Yet, NISQ devices are able to handle smaller quantum problems, which are error-resistent. Hybrid-classical computing is way of structuring problems and using the best of both worlds, by combining quantum and classical approaches. The classical computations are used to reduce the problem complexity and create a task that can be handled by current quantum computers. This way, quantum computing, in the state it currently is, can already be accessed and utilized for real-world applications. 

Multiple decomposition strategies exist, to be able to create a hybrid approach. A problem might be deconstructed into smaller problems, which can then be solved either quantumly or classically. Ways to decompose a problem are by divide and conquer (top-down) or dynamic programming (bottom-up). 

Another possibility is to create algorithms with classical and quantum parts, by identifying building blocks of the algorithm, and replacing possible steps with quantum solutions. Examples are variational techniques, like the variational quantum eigensolver (VQE) or quantum approximate optimization algorithm (QAOA), which combine quantum state preparation and classical optimization.

Programs itself can also by divided by using libraries and pre-created building blocks. It is also important to note the difference of running programs on a simulator or an actual quantum computer, both have pros and cons. The simulator is slower, hence does not provide an actual quantum advantage, but achieves a higher accuracy and is less error-prone. It is therefore interesting for testing out ideas and their potential, before concerning an actual quantum computer.

Another decomposition strategy is to deconstruct the circuits itself. The circuit depth on NISQ devices is limited, which does not make it possible to run quantum circuits with large depths. 

### Variational Techniques (VQE and QAOA)

The variational technique VQE is an important building block of the hybrid quantum toolkit. It is used to estimate the lowest eigenvalue of a physical system, represented by a Hamiltonian H, or a Hermitian matrix. The Hamiltonian is a definition from the field of pysics. Its eigenvalues describe the energy values we can measure in a system, because the Hamiltonian describes possible states of such. This makes VQE useful for applications in quantum mechanics, as finding the lowest eigenvalue resembles finding the ground state of a system. 

VQE is a variational algorithm, because it varies parameters toward the optimization goal. It is called eigensolver, because it aims to find the lowest eigenvalue, or more precisely, the upper bound of the lowest eigenvalue. The separate building blocks of VQE are the following:

1) Initialize the Problem (quantum)
2) Ansatz Preparation (quantum)
3) Evaluate Cost Function (classic)
4) Optimization of Parameters (classic)
5) Update Quantum Circuit Parameters (quantum)

<div>
<img src="https://i.stack.imgur.com/nhC1j.jpg" width="500" height="300"/>
</div>

##### 1. Initialization 

A quantum circuit is first initialized in a state $\ket{\rho}$ that does not use any tunable parameters, for example $X$, $CNOT$ or $H$ gate. The goal for this state is to make the optimization converge faster by being an optimal starting point.

##### 2. Ansatz Preparation 

After creating a reference state in the initialization step, which does not have any parameters, we now apply a parameterized circuit. The parameters of this circuit are later optimized by the classial optimization step. Therefore, the so-called Ansatz is an educated guess with parameters, which are utilized by the optimization algorithm to explore the search space. We can use problem-specific Ansatze to reduce our search space depending on the problem at hand and increase the speed of our algorithm.

##### 3. Evaluate Cost Function

The cost function is used to describe how well we are currently doing regarding our optimization goal. The inputs of the cost function are the measurements of the quantum circuit we created in step 1 and 2. The definition of the cost function depends on our goal and our application. In the quantum computer the cost function is represented by the Hamiltonian. 

##### 4. Optimization of Parameters

The evaluation of the cost function is used to optimize the tunable parameters of the Ansatz in this step. Any classical optimizer can be used to search for the point that minimizes the cost function. The resulting parameters are used to update the Ansatz.

QAOA is another variational quantum algorithm that is used to solve combinatorial optimization problems. As it uses VQE, it is viable on NISQ computers. The idea is to exploit problem structure and by that achieve a quantum speedup for combinatorial NP-complete problems. This is realized by finding the ground state or highest energy eigenstate using a variational ansatz.  

The main structural difference of QAOA compared to VQE is that QAOA uses an alternating ansatz with which we are trying to find the eigenstate that corresponds to the solution of our combinatorial problem. The alternating Ansatz consists of a Cost Hamiltonian $H_C$, which encodes the problem, and a Mixing Hamiltonian $H_M$, which is used to mix the states among each other. The most important attribute of $H_M$ is that it does not commute with $H_C$, because otherwise we are not able to explore all possible solutions. 

<div>
<img src="https://www.researchgate.net/publication/341047640/figure/fig1/AS:886001115361281@1588250535257/Schematic-diagram-of-hybrid-quantum-classical-QAOA-3-QAOA-based-scheme-for-solving-TSP.ppm" width="500" height="300">
</div>

The QAOA circuit consists of following steps:

##### 1) Ansatz Construction

Combine the Hamiltonians $H_C$ and $H_M$ and apply them to the initial state $p$ times. The Ansatz is applied using two parameters $\gamma$ and $\beta$. 

##### 2) Expectation Value

Find the expectation value by evaluating the cost function.

##### 3) Parameter Optimization

Classically optimize the parameters $\gamma$ and $\beta$ to maximize the expectation value. 

##### 4) Obtain Probabilities

Rerun the Ansatz with the optimized parameters.

## Deutsch-Jozsa Algorithm

![image](imgs/deutsch_jozsa.png)
*source: Slides Lecture 10 Deutsch's Algorithm Part 2 Deutsch Jozsa*

### Quantum Advantage
The Deutsch-Jozsa algorithm uses equal superposition by applying $n$ Hadamard gates to $n$ number of qubits. Then, the universal boolean oracle $U_f$ is applied, where 

$U_f \left | x_1x_2...x_ny \right> = \left | x_1x_2...x_n f(x_1x_2...x_n)\oplus y\right>$.

Compared to the binary boolean operator $B_f$ with 

$B_f \left | xy \right> = \left | xf(x) \oplus y \right>$,

$U_f$ takes $n$ number of inputs.

This difference made me realize how the equal superposition state, created by applying a Hadamard gate to each of the $n$ qubits, simulates $2^n$ possible input combinations for the oracle function $f$, and how quantum speed up can be accomplished. 

### Phase Kickback

The Deutsch-Jozsa algorithm uses the principle of phase kickback, which means that operations on one qubit, the auxiliary qubit, may change the phase of other qubits. For example, applying $H$ to the first qubit after applying $U_f$, for $n=2$ qubits, results in following state for a constant oracle function $f$:

$\left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left| f(0) \right> - \left| \overline{f(0)} \right>) = \pm \left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left | 0 \right> - \left | 1 \right>)$

In more detail:

$\left| f(0) \right> - \left| \overline{f(0)} \right> = 
\begin{cases} \left | 0 \right> - \left | 1 \right>, \text{ if} f(0) = 0 \\ \left | 1 \right> - \left | 0 \right>, \text{ if} f(0) = 1 \end{cases}$

Hence, if $f(0) = 0$:

$\left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left| f(0) \right> - \left| \overline{f(0)} \right>) = + \left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left| 0 \right> - \left| 1 \right>)$

If $f(1) = 1$:

$\left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left| f(0) \right> - \left| \overline{f(0)} \right>) = \left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left| 1 \right> - \left| 0 \right>) = \left | 1 \right> \otimes \frac{1}{\sqrt{2}}(-1)(\left| 0 \right> - \left| 1 \right>) = - \left | 1 \right> \otimes \frac{1}{\sqrt{2}}(\left| 0 \right> - \left| 1 \right>)$

The mathematical attribute of the tensor product makes it possible to take a negative phase from one qubit state, and move it to another qubit's mathematical state, as the phase is nothing other than a negative scaler:

given kets $\left | \phi \right>, \left | \psi \right>, \left | \rho \right>$, and $\left | \sigma \right>$, and $r \in \mathbb{R}$, then $r(\left | \phi \right> \otimes \left | \psi \right>) = (r\left | \phi \right>) \otimes \left | \psi \right> = \left | \phi \right> \otimes (r\left | \psi \right>)$

## Relationship of Boolean and Phase Oracle

The boolean oracle $B_f$ outputs either $f(x)$ or $\overline{f(x)}$, depending on the input $y$. The phase oracle $O_f$ tags a target element with a negative phase. Both can be described by using each others definition:

![image](imgs/rel_boolean_phase_oracle.png)
*source: Slides Lecture 10 Deutsch's Algorithm Part 2 Deutsch Jozsa*

Since applying $B_f$ to $n$ qubits, with qubit $q_n$ in state $\ket{-}$, the state changes to $\frac{1}{\sqrt{2}}(\ket{f(x) \oplus 0}-\ket{f(x) \oplus 1})$, according to the mechanism of $B_f$.

We know that $f(x) \oplus 1 = \overline{f(x)}$, therefore we can rewrite it as $\frac{1}{\sqrt{2}}(\ket{f(x)}-\ket{\overline{f(x)}})$.

This lets us distinguish two cases:

$f(x) = 1$, then $\frac{1}{\sqrt{2}}(\ket{1}-\ket{0})$

$f(x) = 0$, then $\frac{1}{\sqrt{2}}(\ket{0}-\ket{1})$

We can rewrite it as 

$(-1)^{f(x)}\ket{-} = \begin{cases}(-1)\ket{-} \\ \ket{-}\end{cases}$

The $(-1)^{f(x)}$ is just a fancy notation for the fact that the ket has a negative phase when $f(x)=1$, because $(-1)^1 = (-1)$, in contrast to when $f(x)=0$, which results in $(-1)^0 = 1$.

Now, we use phase kickback, which we encountered with the Deutsch-Jozsa algorithm. The phase is kicked to the element $\alpha_x$, for which $f(x)=1$, which is exactly how $O_f$ works.

## Changing Basis in Linear Algebra

We can think of vectors $\begin{bmatrix}
  \alpha\\
  \beta
\end{bmatrix}$ as scalars, where $\alpha$ scales the basis vector $\begin{bmatrix}
  1\\
  0
\end{bmatrix}$ and $\beta$ scales $\begin{bmatrix}
  0\\
  1
\end{bmatrix}$. The choice of the basis vectors and unit of distance results in the coordinate system. Following the same idea, changing the basis vectors can be seen as an alternate way of describing the same vectors in a modified coordinate system. It can be thought of a differently shaped grid. A translation between two different grids is done by multiplying each coordinate of the vector to be translated $\begin{bmatrix}
  \alpha\\
  \beta
\end{bmatrix}$ with the basis vectors of the source grid and summing the result. This is equal to multiplying the matrix, with the basis vectors of the new grid system as columns, $\begin{bmatrix}
  x_1 & y_1\\
  x_2 & y_2
\end{bmatrix}$, with the vector to be translated.

## Eigenvalues and Eigenstates in Quantum Computing

Given a unitary matrix $U$:

\begin{equation}
U\ket{\psi} = \lambda\ket{\psi}
\end{equation}

, with eigenvalue $\lambda$ corresponding to eigenvector $\ket{\psi}$. Since all quantum gates are unitary matrices, and unitary matrices have eigenvalues with a unit norm, the eigenvalue can be written in the form $\lambda = e^{2\pi i\theta}$. These eigenvalues have a magnitude of $1$, which means that their value will be between $0$ and $1$.

The eigenvectors of a unitary form an orthonormal set, why we can write $U$ as a *spectral decomposition*, which means as a function of its eigenvalues and eigenvectors:

\begin{equation}
U = \sum_\lambda e^{2\pi i\theta_\lambda}\ket{\psi_\lambda}\bra{\psi_\lambda}
\end{equation}

This is also the reason why for any diagonal matrix, the eigenvalues can be found on the diagonal.
